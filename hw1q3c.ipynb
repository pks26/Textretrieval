For different values of d, I initialised the mean and covariance matrix according to the values of d. Mean matrix is initialised randomly, while covariance matrix is initialised to an identity matrix.

```
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
d = 50
m = np.random.randint(2, size=d)
c = np.identity(d)
z = np.random.multivariate_normal(mean=m, cov=c, size=10000)
y = np.transpose(z)
sns.jointplot(x=y[0],
              y=y[1], 
              kind="kde", 
              space=0);
```

Plotting the histogram, standard deviation and average

```
import matplotlib.pyplot as plt
plt.hist(z,)
plt.xlabel('Samples')
plt.ylabel('Freq')
plt.title('Histogram of samples')
plt.grid(True)
plt.show()
```

```
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

import math

mu = 0
variance = 1
sigma = math.sqrt(variance)
z = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(z, stats.norm.pdf(z, mu, sigma))
plt.grid(True)
plt.show()
```
