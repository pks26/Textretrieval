The function computeTF computes the TF score for each word, by document

```
def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict


tf1 = computeTF(wordDict1, w1)
tf2 = computeTF(wordDict2, w2)
print(tf1)
```

The function computeIDF computes the IDF score of every word.

```
import math

def computeIDF(docList):
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / (float(val)+ 0.01)) 
        # added 0.01 in the denominator, hence some idfs are negative(should be 0 ideally)
    return idfDict

idfs = computeIDF([wordDict1, wordDict2])
print(idfs)
```

The function computeTFIDF below computes the TF-IDF score for each word, by multiplying the TF and IDF scores.

```
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf

tfidf1 = computeTFIDF(tf1, idfs)
tfidf2 = computeTFIDF(tf2, idfs)

import pandas as pd
pd.DataFrame([tfidf1, tfidf2])
```
